```python
from langchain.chat_models import init_chat_model
from langchain_core.tools import tool, StructuredTool
from langchain_core.prompts import ChatMessagePromptTemplate, ChatPromptTemplate
from mistralai import Mistral
from dotenv import load_dotenv
from langgraph.store.memory import InMemoryStore
from typing import Optional, List, Dict
from typing_extensions import Annotated, TypedDict
from langgraph.prebuilt import create_react_agent
from agent_tools import PreprocessingTool
import os
```

```python
class ExtractKeywords(TypedDict):
    keywords: Annotated[str, ..., 'Keywords or concepts listed by the authors (under Unified Astronomy Thesaurus Concepts, Keywords, etc)']
```

```python
data_agent_tools = PreprocessingTool()
get_text_prompt = ChatPromptTemplate.from_messages(
    [
        ('system', 'Extract the contents of the pdf file given.'),
        ('user', '{file}')
    ]
)

llm = init_chat_model('mistral-large-latest', model_provider='mistralai')
```

```python
agent = create_react_agent(llm, data_agent_tools.get_tools())
```

```python
dirs = '../SSWLAB_논문'
files = os.listdir(dirs)

file = files[0]
file = os.path.join(dirs, file)
file
```

```python
result = agent.invoke({'messages': ('human', f"Extract all content from this file exactly: {file}")})
```

```python
paper = result['messages'][-1].content
```
